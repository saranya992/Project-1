```python
import pandas as pd
import glob
df = pd.read_csv("ap_bus_details.csv")
df = pd.read_csv("assam_bus_details.csv")
df = pd.read_csv("chandigarh_bus_details.csv")
df = pd.read_csv("himachal_bus_details.csv")
df = pd.read_csv("kaac_bus_details.csv")
df = pd.read_csv("kerala_bus_details.csv")
df = pd.read_csv("rajasthan_bus_details.csv")
df = pd.read_csv("sb_bus_details.csv")
df = pd.read_csv("Telangana_bus_details.csv")
df = pd.read_csv("up_bus_details.csv")
df = pd.read_csv("wb_bus_details.csv")
```

```python
# List of CSV file paths
csv_files = [
    "ap_bus_details.csv", "assam_bus_details.csv", "chandigarh_bus_details.csv",
    "himachal_bus_details.csv", "kaac_bus_details.csv", "kerala_bus_details.csv",
    "rajasthan_bus_details.csv", "sb_bus_details.csv", "Telangana_bus_details.csv",
    "up_bus_details.csv", "wb_bus_details.csv"
]   

# Read each CSV file into a DataFrame and store it in a list
df_list = [pd.read_csv(file) for file in csv_files]

# Concatenate all DataFrames in the list
combined_df = pd.concat(df_list, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv("all_bus_routes.csv", index=False)
```

```python
# Add an id column
id_column = pd.Series(range(1, len(combined_df) + 1), name='id')

# Add the id column to the DataFrame
combined_df = pd.concat([id_column, combined_df], axis=1)
```

```python
#convert datatype into based on our need
combined_df['Departing_Time']=pd.to_datetime(combined_df['Departing_Time'],format="%H:%M").dt.strftime("%H:%M")
combined_df['Reaching_Time']=pd.to_datetime(combined_df['Reaching_Time'],format="%H:%M").dt.strftime('%H:%M')
```

```python
combined_df['Departing_Time']
combined_df.info()
```

```python
# Convert all entries to strings
combined_df['Star_Rating'] = combined_df['Star_Rating'].astype(str)

# Split the values at newline characters and take the first part
combined_df['Star_Rating'] = combined_df['Star_Rating'].str.split('\n').str[0]

# Convert the values to numeric, replacing non-convertible values with NaN
combined_df['Star_Rating'] = pd.to_numeric(combined_df['Star_Rating'], errors='coerce')
```

```python
combined_df['Star_Rating'] 
```

```python
combined_df['Seat_Availability']
```

```python
# Clean and convert Seat_Availability column
combined_df['Seat_Availability'] = (combined_df['Seat_Availability']
                                    .astype(str)                        # Step 1: Convert to string
                                    .str.strip()                        # Step 2: Strip whitespace
                                    .str.replace(" Seats available", " ", regex=False)  # Step 3: Remove text
                                    .str.strip())                       # Step 2 (optional): Strip again after replacing
                                    
# Step 4: Convert cleaned values to numeric, then to integers
combined_df['Seat_Availability'] = pd.to_numeric(combined_df['Seat_Availability'], errors='coerce').fillna(0).astype(int)

# Final result
print(combined_df['Seat_Availability'].head())
```

```python
from decimal import Decimal

# Step 1: Ensure 'Price' column is of string type and then remove 'INR '
combined_df['Price'] = combined_df['Price'].astype(str).str.replace('INR ', '', regex=False)

# Step 2: Convert the cleaned 'Price' column to Decimal and round to 2 decimal places
combined_df['Price'] = combined_df['Price'].apply(lambda x: round(Decimal(x), 2))

# Final result
print(combined_df['Price'].head())
```

```python
cursor.execute("USE Redbus_Project")

# Drop any existing table and create a new one based on the DataFrame schema
combined_df = pd.read_csv("bus_routes.csv")
combined_df = combined_df.dropna()  # Drop null values

cursor.execute("""
CREATE TABLE IF NOT EXISTS bus_routes (
    id INT AUTO_INCREMENT PRIMARY KEY,
    Route_Name VARCHAR(255),
    Route_Link VARCHAR(255),
    Bus_Name VARCHAR(255),
    Bus_Type VARCHAR(255),
    Departing_Time VARCHAR(255),
    Duration VARCHAR(255),
    Reaching_Time VARCHAR(255),
    Star_Rating FLOAT,
    Price DECIMAL(10, 2),
    Seat_Availability INT
);
""")
```

```python
combined_df.info()
```

```python
sql_query = """
        INSERT INTO bus_routes (

            Route_Name, Route_Link, Bus_Name, Bus_Type, Departing_Time, Duration, 
            Reaching_Time, Star_Rating, Price, Seat_Availability
        ) VALUES (
            %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
        );
        """
        
        # Iterate over DataFrame rows
for index, row in combined_df.iterrows():
            
            # Execute the query with parameters
            cursor.execute(sql_query, (
                row['Route_Name'],
                row['Route_Link'],
                row['Bus_Name'],
                row['Bus_Type'],
                row['Departing_Time'],
                row['Duration'],
                row['Reaching_Time'],
                row['Star_Rating'],
                row['Price'],
                row['Seat_Availability'],
))
mydb.commit()   
cursor.close()
mydb.close()         
```
